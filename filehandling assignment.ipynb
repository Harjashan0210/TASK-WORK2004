```python
name=[]
marks=[]

with open("example.txt") as file2:
    for line in file2:
        data=line.strip().split(",")
        name=data[0]
        marks=int(data[1])

        if marks>=45: 
            print(name,marks ,"- Passed")
        else:
            print(name,marks,"_failed")
```

```python
count=0

with open("salary.txt") as file:
    for line in file:
        data=line.strip().split(",")
        name=data[0]
        salary=int(data[1])

        if salary>=50000:
            print(name,salary,"Satisfied")
            count+=1
        else:
            print(name,salary,"Not Satisfied")

print("The Total Number of Employees is satisfied =",count)

```
