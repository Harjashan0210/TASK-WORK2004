```python
# Read daily temperature data from a CSV file using Pandas. 
# Use NumPy to calculate the highest, lowest, and average temperatures.
# Write functions to classify each day as:


df2=pd.read_csv("temperature_data.csv")
df2
```

```python
a=np.array(df2["Temperature"])
high=a.max()
cold=a.min()
moderate=a.mean()
print("Highest = ",high)
print("Cold = ",cold)    
print("moderate = ",moderate)

def calculate():
    if a<30:
        print("Cold")
    elif a<=40:
        print("Moderate")
    else:
        print("HOT")


df2['Report']=df2['Temperature'].apply(calculate)
print(df2)
```
